```python
#Collect the datetime of the posts were created
python = []
tensorflow = []
keras = []
machine_learning = []
deep_learning = []
nlp = []
neural_network = []

for index, row in dataset.iterrows():
    if ("python" in str(row["title"])):
            python.append(row['creation_date'])
    elif ("tensorflow" in str(row["title"])):
            tensorflow.append(row['creation_date'])
    elif ("keras" in str(row["title"])):
            keras.append(row['creation_date'])
    elif ("machine-learning" in str(row["title"]) or "machine learning" in str(row["title"]) or "ml" in str(row["title"])):
            machine_learning.append(row['creation_date'])
    elif ("deep-learning" in str(row["title"]) or "deep learning" in str(row["title"]) or "deep neural net" in str(row["title"])):
            deep_learning.append(row['creation_date'])
    elif ("nlp" in str(row["title"]) or "natural language processing" in str(row["title"])):
            nlp.append(row['creation_date'])
    elif ("neural-network" in str(row["title"]) or "neural network" in str(row["title"]) or "mlp" in str(row["title"])):
           neural_network.append(row['creation_date'])
```
